## The following code does the following:
### a) Merges all data into a single dataframe
### b) Calculates raw highest rated movie (no conditionals)

```python
# merge all data into one dataframe
rat_and_use=pd.merge(ratings,users,on='user_id')
all_together=pd.merge(rat_and_use,movies,on='movie_id')
all_together.tail()
```

```python
# here I am finding the highest rated movie with no conditionals

# temp value for comparison
aveRating = 0.0
# save movie_id of highest rated movie
highestRated = 0

# iterate through every movie in list
for x in range(0, all_together.movie_id.max()+1):
    # grab current movie in iteration
    tempList = all_together.loc[all_together['movie_id'] == x]
    
    # if rating of this movie is higher, this movie is the new highest rated movie
    if tempList.rating.mean() > aveRating:
        aveRating = tempList.rating.mean()
        highestRated = x

print(highestRated)
print(all_together.loc[all_together['movie_id'] == highestRated].head())
# The Gate of Heavenly Peace
```

## The following code calculates the highest rating among males and females (calculated separately then combined) with
### a) at least 10 ratings
### b) within the genres 'Comedy', 'Adventure', and/or 'Romance' (or some combination)

```python
# just seeing what kinds of genres I am dealing with
all_together.genres.unique()
```

```python
# pluck out all of the fitting genres
chosenGenres = all_together.loc[all_together['genres'].str.contains('Comedy' and 'Adventure' and 'Romance')]

# divide this dataframe into two dataframes; one male, one female
# (this adjusts for an uneven male to female ratio)
males = chosenGenres.loc[chosenGenres['gender'] == 'M']
females = chosenGenres.loc[chosenGenres['gender'] == 'F']
```

```python
chosenGenres.head()
```

```python
males.head()
```

```python
females.head()
```

```python
len(chosenGenres[chosenGenres['movie_id'] == 914].rating)
```

```python
# temporary variable for comparison within loop; used to compare current movie with the current highest movie
aveCombined = 0.0
# saves movie_id of highest rated movie within my conditionals
highestRatedBoth = 0

# cycle through all movies
for x in range(0, chosenGenres.movie_id.max()+1):
    # grab next movie for both males and females
    maleTemp = males.loc[males['movie_id'] == x]
    femaleTemp = females.loc[females['movie_id'] == x]
    # if there are at least 10 ratings
    if len(chosenGenres[chosenGenres['movie_id'] == x].rating) >= 10:
        # lets find the highest rating for males and females, calculated separately then combined
        if maleTemp.rating.mean() + femaleTemp.rating.mean() > aveCombined:
            aveCombined = maleTemp.rating.mean() + femaleTemp.rating.mean()
            highestRatedBoth = x

print(highestRatedBoth)
print(all_together.loc[all_together['movie_id'] == highestRatedBoth].head())
# City Lights, which is indeed a cute, funny, romantic movie!
```

## The following code does the following:
### a) defines and creates a new column for age ranges
### b) breaks down survival ratios to include age range (so age range, class and sex are all included in the breakdown)

```python
# first, divide data into age ranges
child = t_df.loc[t_df['age'] < 14]
teen = t_df.loc[t_df['age'] >= 14]
teen = teen.loc[teen['age'] < 21]
adult = t_df.loc[t_df['age'] >= 21]
adult = adult.loc[adult['age'] < 65]
senior = t_df.loc[t_df['age'] >= 65]
```

```python
# create age range column
t_df["age_range"] = ""

# loop through to fill out new column properly
for x in range(1, len(t_df)+1):
    if (t_df.age[x] < 14.0):
        t_df.age_range[x] = "child"
    elif (t_df.age[x] >= 14.0) and (t_df.age[x] < 21.0):
        t_df.age_range[x] = "adolescent"
    elif (t_df.age[x] >= 21.0) and (t_df.age[x] < 65):
        t_df.age_range[x] = "adult"
    elif (t_df.age[x] >= 65.0):
        t_df.age_range[x] = "senior"
    else:
        t_df.age_range[x] = "NaN"
```
